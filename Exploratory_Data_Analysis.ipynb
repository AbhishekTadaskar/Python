# EDA :- Exploratory Data Analysis
It means:
Looking at your data before building any model â€” to understand it, clean it, and find patterns.

ðŸŽ¯ Why EDA is important:
1.   Helps you understand your data
2.   Prepares data for machine learning
3.   Helps avoid wrong conclusions











```python
import pandas as pd

df=pd.read_csv('/content/car_price_dataset.csv')
df
```

```python
# 1. What is the average price of all cars in the dataset?
p=df['Price'].mean()
p
```

```python
# 2. Which car brand has the highest average price?
df[df['Price']== df['Price'].max()]['Brand'][1100]
```

```python
# 3. What is the most common fuel type in the dataset?
f=df['Fuel_Type'].mode()[0]
f
```

```python
# 4. How many cars have a mileage greater than 200,000?
m=df[df['Mileage']>200000].value_counts(df['Brand']).sum()
m
```

```python
# 5. What is the distribution of cars by transmission type?
df['Transmission'].value_counts()
```

```python
# 6. Which car model has the highest price?
highest = df.loc[df['Price'].idxmax(), ['Model', 'Price']]
highest
```

```python
# 7. What is the average price of cars by fuel type?
df.groupby('Fuel_Type')['Price'].mean()
```

```python
# 8. How many cars are from the year 2020 or later?
count=df[df['Year']>= 2020].count()
count.values[0]
```

```python
# 9. What is the average engine size for each brand?
average_engine_size_by_brand = df.groupby('Brand')['Engine_Size'].mean()
average_engine_size_by_brand
```

```python
# 10. Which car has the lowest mileage?
lowest_mileage_car = df.loc[df['Mileage'].idxmin(), [ 'Model']]
print(lowest_mileage_car)
```

```python
# 11. What is the correlation between mileage and price?
df['Mileage'].corr(df['Price'])
```

```python
# 12. How many cars have more than 2 owners?
c=df[df['Owner_Count']>2].count().values[0]   #or replace .values =.iloc[0]
c
```

```python
# 13. What is the average price of cars by year?
# df['Price'].mean()
df.groupby('Year')['Price'].mean()
```

```python
# 14. Which car has the highest number of doors?
df[df["Doors"] == df["Doors"].max()]
```

```python
# 15. What is the average mileage for each fuel type?
df.groupby('Fuel_Type')['Mileage'].mean()
```

```python
# 16. How many cars are from the brand 'Toyota'?
df[df['Brand']=='Toyota'].count().values[0]
```

```python
# 17. What is the average price of cars with automatic transmission?
# df.groupby('Transmission')['Price'].mean().values[0]    or
T = df[df["Transmission"] == "Automatic"]
T = T["Transmission"]
df.groupby(T)["Price"].mean()
```

```python
# 18. Which car has the highest owner count?
df[df['Owner_Count']==df['Owner_Count'].max()]
```

```python
# 19. What is the average engine size for cars with diesel fuel type?
disel = df[df["Fuel_Type"] == "Diesel"]
disel = disel["Fuel_Type"]
df.groupby(disel)["Engine_Size"].mean()
```

```python
# 20. How many cars have a price greater than $10,000?
df[df['Price']>10000].count()[0]
```
