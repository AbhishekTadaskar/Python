# Lists

```
# ðŸ“‹ Python Lists - Detailed Explanation with Examples

# A list is a collection which is ordered and **mutable** (can be changed).
# Lists are defined using square brackets []

# âœ… Creating a list
fruits = ["apple", "banana", "cherry"]
print("Original list:", fruits)

# âœ… Accessing elements (index starts from 0)
print("First fruit:", fruits[0])

# âœ… Modifying elements
fruits[1] = "blueberry"
print("After modification:", fruits)

# âœ… Adding elements
fruits.append("orange")
print("After appending:", fruits)

# âœ… Inserting at specific position
fruits.insert(1, "mango")
print("After inserting:", fruits)

# âœ… Removing elements
fruits.remove("apple")
print("After removing apple:", fruits)

# âœ… Popping last item
last_item = fruits.pop()
print("Popped item:", last_item)

# âœ… Iterating through list
for fruit in fruits:
    print("Fruit:", fruit)

# âœ… List slicing
print("Sliced list [1:3]:", fruits[1:3])

# âœ… List comprehension (creating new list with logic)
squares = [x**2 for x in range(5)]
print("List of squares:", squares)
```

# Dictionaries

```
# ðŸ“š Python Dictionaries - Detailed Explanation with Examples

# A dictionary is an unordered collection of key-value pairs.
# Defined using curly braces {} with key: value

# âœ… Creating a dictionary
person = {
    "name": "Alice",
    "age": 25,
    "city": "Pune"
}
print("Original dictionary:", person)

# âœ… Accessing values
print("Name:", person["name"])

# âœ… Modifying values
person["age"] = 30
print("After modifying age:", person)

# âœ… Adding new key-value pairs
person["email"] = "alice@example.com"
print("After adding email:", person)

# âœ… Removing a key
del person["city"]
print("After removing city:", person)

# âœ… Looping through dictionary
for key, value in person.items():
    print(f"{key}: {value}")

# âœ… Checking key existence
print("Is 'name' a key?", "name" in person)
```

# Tuples

```
# ðŸŸ° Python Tuples - Detailed Explanation with Examples

# A tuple is an ordered and **immutable** collection.
# Defined using parentheses ()

# âœ… Creating a tuple
colors = ("red", "green", "blue")
print("Original tuple:", colors)

# âœ… Accessing elements
print("First color:", colors[0])

# âœ… Tuple unpacking
r, g, b = colors
print(f"R: {r}, G: {g}, B: {b}")

# âœ… Iterating through tuple
for color in colors:
    print("Color:", color)

# âœ… Tuples can be used as dictionary keys (because they are immutable)
location = {(10.5, 20.3): "Home"}
print("Location dict:", location)
```

# Sets

```
# ðŸ§µ Python Sets - Detailed Explanation with Examples

# A set is an unordered collection of **unique** elements.
# Defined using curly braces {}, like dictionaries but without key:value

# âœ… Creating a set
numbers = {1, 2, 3, 2, 1}
print("Original set (duplicates removed):", numbers)

# âœ… Adding elements
numbers.add(4)
print("After adding 4:", numbers)

# âœ… Removing elements
numbers.discard(2)  # discard doesn't raise error if element not found
print("After discarding 2:", numbers)

# âœ… Set operations
odds = {1, 3, 5}
evens = {2, 4, 6}
print("Union:", odds.union(evens))
print("Intersection:", odds.intersection(numbers))
print("Difference:", numbers.difference(odds))

# âœ… Checking membership
print("Is 3 in numbers?", 3 in numbers)

# âœ… Iterating over set
for num in numbers:
    print("Number:", num)
```
